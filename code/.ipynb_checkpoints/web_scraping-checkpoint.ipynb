```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
data_img_folder = "../datasets"
```

```python
# Function to scrape images and labels from a single page
def scrape_page(url):
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Failed to fetch {url}")
        return

    soup = BeautifulSoup(response.content, 'html.parser')

    # Find image containers
    image_containers = soup.find_all(image_containers = soup.find_all('li', class_='layout-grid layout-metro column_4 active')
    print(image_containers)

                                     
    # Extract image URLs and labels
    #for container in image_containers:
        #image_tag = container.find('img')
        #image_url = image_tag['src']
        #label = image_tag['alt']
        
        # Download image and save with label
        #download_image(image_url, label)
        #print(image_url)

# Function to download image and save with label
def download_image(url, label):
    image_data = requests.get(url).content
    filename = f"{label}.jpg"
    with open(os.path.join('images', data_img_folder), 'wb') as f:
        f.write(image_data)
    print(f"Image saved: {data_img_folder}")
```

```python
# Main function to iterate through pages and scrape
def main():
    base_url = "https://liansenghin.com.sg/product-category/tiles/"
    page_number = 1

    # Create a directory to store images
    os.makedirs(os.path.join('images', data_img_folder), exist_ok=True)

    # Iterate through pages
    while True:
        page_url = f"{base_url}page/{page_number}/"
        response = requests.get(page_url)
        if response.status_code != 200:
            print(f"No more pages. Exiting.")
            break

        print(f"Scraping page {page_number}...")
        scrape_page(page_url)
        #page_number += 1
```

```python
if __name__ == "__main__":
    main()
```
